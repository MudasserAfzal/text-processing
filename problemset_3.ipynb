## Exercise 1 Tokenization NLTL

```python
!pip install nltk
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import nltk

nltk.download('punkt')
```

```python
text = """
Artificial Intelligence (AI) is a rapidly evolving field that is transforming industries and daily life.
It enables machines to learn from data, make decisions, and solve problems with minimal human intervention.
AI applications are seen in healthcare, finance, automotive, and many other sectors.
As technology advances, the potential of AI continues to grow, raising questions about ethics, safety, and societal impact.
"""
```

```python
tokens = word_tokenize(text)
print("List of tokens:", tokens)
num_tokens = len(tokens)
print("Number of tokens:", num_tokens)
```

```python
frequency_dist = FreqDist(tokens)

# Print the frequency distribution of each token
print("Frequency of each token:")
for word, frequency in frequency_dist.items():
    print(f"{word}: {frequency}")
```

## Exercise 2 WORD cloud

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/JaneEyre.txt"
with open(file_path, 'r') as file:
    text = file.read()
```

### The general wordcloud. The common words here in the text file are

1.   now
2.   little
3.   well
4.   will
5.   said
6.   rochester
7.   thought



```python
wordcloud = WordCloud(
width=800,
height=400,
background_color="white",
stopwords=set(STOPWORDS),
min_font_size=10,
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

### Using different shapes to draw cloud

```python
image1 = "/content/drive/MyDrive/Colab Notebooks/hiyy.png"
image2 = "/content/drive/MyDrive/Colab Notebooks/img1.png"
```

```python
from PIL import Image
import numpy as np

mask1 = Image.open(image1)
mask1 = mask1.convert("L")  # Convert to grayscale (L mode)
mask1 = np.array(mask1)
mask2 = Image.open(image2)
mask2 = mask2.convert("L")  # Convert to grayscale (L mode)
mask2 = np.array(mask2)
```

```python
wordcloud = WordCloud(
    width=800,
    height=800,
    background_color="white",
    stopwords=STOPWORDS,
    mask=mask1,
    min_font_size=5,
).generate(text)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
wordcloud = WordCloud(
    width=800,
    height=800,
    background_color="white",
    stopwords=STOPWORDS,
    mask=mask2,
    min_font_size=10,
).generate(text)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
